# Modelacion
Modelo de regresion lineal

Para trabajar los casos de modelo de regresion lineal se deben seguir los siguientes pasos:
1. Plantear el modelo con su formula
2. Ajustar el modelo utilizando el metodo .fit()
3. Mostrar los resultados del modelo

```python
#Se importa libreria para los modelos
import statsmodels.formula.api as smf
```

```python
# Cual es la relacion entre el CO2 y el PIB de cada pais
modelo_1 = smf.ols('gdp ~ co2', data= df_limpia)
```

```python
modelo_1= modelo_1.fit()
```

```python
modelo_1.summary()
```
